```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Define a 'Loss Function' (e.g., Portfolio Error)
# f(x, y) = x^2 + y^2 (The bottom is 0,0)
def loss_function(w):
    return w[0]**2 + w[1]**2

# 2. Define the Gradient (The Jacobian/Derivative)
# df/dx = 2x, df/dy = 2y
def gradient(w):
    return np.array([2*w[0], 2*w[1]])

# 3. Gradient Descent Loop (The AI 'Learning' Process)
w = np.array([4.0, 4.0]) # Starting high
learning_rate = 0.1
history = []

for i in range(20):
    history.append(w.copy())
    grad = gradient(w)
    w = w - learning_rate * grad # Moving opposite to the gradient

# 4. Visualize the "Descent" to the optimal point
history = np.array(history)
plt.plot(history[:,0], history[:,1], 'ro-')
plt.plot(0, 0, 'g*', markersize=15, label="Optimal Model")
plt.title("Day 5: Visualizing Gradient Descent")
plt.legend()
plt.show()
```
