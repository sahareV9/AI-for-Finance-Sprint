```python
import numpy as np
import time

# Create a 2000x2000 matrix (simulating a large portfolio)
size = 2000
A = np.random.rand(size, size)
b = np.random.rand(size)

# Method 1: Matrix Inversion (The "Student" way)
start = time.time()
x_inv = np.linalg.inv(A).dot(b)
print(f"Inversion Time: {time.time() - start:.4f}s")

# Method 2: np.linalg.solve (The "HCL Pro" way - uses LU Decomposition)
start = time.time()
x_solve = np.linalg.solve(A, b)
print(f"Solve Time: {time.time() - start:.4f}s")

```

```python
import numpy as np
import time
import matplotlib.pyplot as plt

sizes = [100, 500, 1000, 2000, 3000]
times = []

for n in sizes:
    A = np.random.rand(n, n)
    b = np.random.rand(n)

    start = time.time()
    np.linalg.solve(A, b) # LU Decomposition
    times.append(time.time() - start)

plt.plot(sizes, times, 'o-', label="Actual Execution Time")
plt.xlabel("Matrix Size (n)")
plt.ylabel("Time (seconds)")
plt.title("The Cubic Wall: O(n^3) Scaling in Financial Modeling")
plt.grid(True)
plt.show()
```
