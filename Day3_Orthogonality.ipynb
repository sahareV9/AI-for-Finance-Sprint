```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Create 'Messy' Finance Data (Simulating a Stock Trend)
# x = time steps, y = stock price with noise
np.random.seed(42)
x = np.array([1, 2, 3, 4, 5, 6])
y = np.array([10.2, 11.5, 12.1, 14.8, 15.2, 17.5])

# 2. Prepare Matrix A (include a column of 1s for the Intercept)
A = np.vstack([x, np.ones(len(x))]).T

# 3. Apply the Normal Equation: (A^T * A)x = A^T * b
# This is the "Best Fit" projection math from MIT Lec 16
ATA = A.T @ A
ATb = A.T @ y
w = np.linalg.solve(ATA, ATb) # Slope and Intercept

# 4. Predict and Visualize
y_pred = A @ w
plt.scatter(x, y, color='blue', label='Actual Citi Data (Simulated)')
plt.plot(x, y_pred, color='red', label='Optimal Projection (Trend)')
plt.legend()
plt.title("Day 3: Linear Regression via Orthogonality")
plt.show()
```
