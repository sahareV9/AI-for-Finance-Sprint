```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Representing the Finance Problem in Matrix Form Ax = b
# A = [[1, 1], [0.05, 0.12]]
# b = [10000, 1000]
A = np.array([[1, 1], [0.05, 0.12]])
b = np.array([10000, 1000])

# 2. Solve for x (Gold) and y (Equity)
shares = np.linalg.solve(A, b)

print(f"Invest in Gold: ₹{shares[0]:.2f}")
print(f"Invest in Equity: ₹{shares[1]:.2f}")

# 3. Visualising the "Row Picture" (Where the lines intersect)
x_vals = np.linspace(0, 10000, 100)
y1 = 10000 - x_vals # From Eq 1
y2 = (1000 - 0.05 * x_vals) / 0.12 # From Eq 2

plt.figure(figsize=(8, 5))
plt.plot(x_vals, y1, label="Total Capital Constraint")
plt.plot(x_vals, y2, label="Target Return Constraint")
plt.plot(shares[0], shares[1], 'ro', label="Optimal Allocation")
plt.xlabel("Gold Investment (₹)")
plt.ylabel("Equity Investment (₹)")
plt.title("Day 1: Portfolio Intersection")
plt.legend()
plt.grid(True)
plt.show()
```
